```python
import glob

import numpy as np
import matplotlib.pyplot as plt
import skimage.io

from src.utils import down_sample
```

```python
glob_path = r'data\healthy\*.jpg'
```

```python
ims_ds = []
ims_gt = []

for file in glob.glob(glob_path):
    im = skimage.io.imread(file)
    im = im.astype(np.float32) / 255
    ims_gt.append(im)
    ims_ds.append(down_sample(im))
```

```python
fig, axs = plt.subplots(4,3,figsize=(20,15))
for i, ax in enumerate(axs):
    ax[0].imshow(ims_gt[i])
    ax[0].set_title("")
    ax[1].imshow(ims_ds[i])
    ax[2].imshow(ims_ds[i][500:600, 500:600])
plt.show()
```
